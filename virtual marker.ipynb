```python
import cv2
import numpy as np
mypoints = []
colours = [[ 0 , 167 , 132 ,  255 , 255 , 255 ],[ 76 , 133 , 119 ,  255 , 255 , 255],[56 , 68 , 80, 92 , 255 , 255],[ 161 , 95 , 65,255 , 255 , 255 ]]
draw = [[255,0,0],[0,255,0],[0,0,255]]
img = np.ones((520,520,3),np.uint8)*255

def drawcontour(mask,frame,clr):
    npoints = []
    Ret,thresh = cv2.threshold(mask,254,255,cv2.THRESH_BINARY)
    _,contour,hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    x=0
    y=0
    for cnt in contour:
        area = cv2.contourArea(cnt)
        if area>= 100:
            cv2.drawContours(frame,cnt,-1,(0,0,255),5)
            perimeter = cv2.arcLength(cnt,True)
            approx = cv2.approxPolyDP(cnt,0.002*perimeter,True)
            x,y,w,h = cv2.boundingRect(approx)
            cv2.circle(frame,(x,y),5,(255,0,0),-1)
        if x!=0 and y!=0:
            npoints.append([x,y])
    
    return npoints,clr

def drawcircle(mypoints,clr):
    if clr == [ 0 , 167 , 132 ,  255 , 255 , 255 ]:
        for point in mypoints:
            cv2.circle(img,(point[0],point[1]),10,(0,0,255),-1)
        #print(i)
    if clr ==[ 76 , 133 , 119 ,  255 , 255 , 255]:
        for point in mypoints:
            cv2.circle(img,(point[0],point[1]),10,(255,0,0),-1)
    if clr ==[55 , 46 , 31, 86 , 124 , 88]:
        for point in mypoints:
            cv2.circle(img,(point[0],point[1]),10,(0,255,0),-1)
    cv2.imshow('output',img)
    
    #print(clr)
    return 0

def main():
    cap = cv2.VideoCapture(0)
    while cap.isOpened():
        ret,frame = cap.read()
        frame = cv2.flip(frame,1)
        
        hsv = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
        for clr in colours:
            low = np.array(clr[0:3])
            high = np.array(clr[3:6])
            mask = cv2.inRange(hsv,low,high)
            newpoints,col = drawcontour(mask,frame,clr)
            if len(newpoints) != 0:
                for npts in newpoints:
                    mypoints.append(npts)
            pp = drawcircle(mypoints,col)
 
            
        
        
        cv2.imshow('live',frame)
        #cv2.imshow('out',mask)
        if cv2.waitKey(1) == 27:
            break
    cv2.destroyAllWindows()
    cap.release()
if __name__ == '__main__':
    main()
```
